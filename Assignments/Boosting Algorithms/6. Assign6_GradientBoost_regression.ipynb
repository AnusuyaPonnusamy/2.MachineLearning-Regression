```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
dataset
```

```python
dataset=dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(independent,dependent, test_size=0.20, random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
regressor=GradientBoostingRegressor(loss='absolute_error', learning_rate=0.5, n_estimators=500, min_samples_split=2, min_samples_leaf=10, max_depth=10,random_state=0)
regressor=regressor.fit(X_train, Y_train)
```

```python
regressor
```

```python
Y_pred=regressor.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, Y_pred)
r_score
```

```python
from sklearn.ensemble import GradientBoostingRegressor
regressor=GradientBoostingRegressor(random_state=0)
regressor=regressor.fit(X_train, Y_train)
Y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, Y_pred)
r_score
```

```python
n_estimators=regressor.n_estimators_ # The number of estimators as selected by early stopping (if n_iter_no_change is specified). Otherwise it is set to n_estimators.

n_trees_per_iteration= regressor.n_trees_per_iteration_ # The number of trees that are built at each iteration. For regressors, this is always 1.

feature_importance= regressor.feature_importances_ # The impurity-based feature importances.

#oob_improvement=regressor.oob_improvement_ #The improvement in loss on the out-of-bag samples relative to the previous iteration. oob_improvement_[0] is the improvement in loss of the first stage over the init estimator. Only available if subsample < 1.0.
# AttributeError: 'GradientBoostingRegressor' object has no attribute 'oob_improvement_'

#oob_scores=regressor.oob_scores_ #The full history of the loss values on the out-of-bag samples. Only available if subsample < 1.0.
# AttributeError: 'GradientBoostingRegressor' object has no attribute 'oob_scores_'

# oob_score= regressor.oob_score_ # The last value of the loss on the out-of-bag samples. It is the same as oob_scores_[-1]. Only available if subsample < 1.0.
# AttributeError: 'GradientBoostingRegressor' object has no attribute 'oob_score_'

train_score=regressor.train_score_ # The i-th score train_score_[i] is the loss of the model at iteration i on the in-bag sample. If subsample == 1 this is the loss on the training data.

init=regressor.init_ # The estimator that provides the initial predictions. Set via the init argument.

estimators=regressor.estimators_ #The collection of fitted sub-estimators.

n_features_in= regressor.n_features_in_ # Number of features seen during fit.

feature_names_in=regressor.feature_names_in_ #Names of features seen during fit. Defined only when X has feature names that are all strings.

max_features=regressor.max_features_ # The inferred value of max_features.
```
