```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[[ 'charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
# CASE 1: Preprocess input from training/test data set and the user input (preinput variable)
```

```python
# Preprocess only input training and test data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=1000)
regressor.fit(X_train, Y_train)
```

```python
Y_Pred=regressor.predict(X_test)
Y_Pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, Y_Pred)
r_score
```

```python
# Preprocessing the user input before predicting the output of the user input
# Note, we have not preprocess the output for user input
preinput=sc.transform([[52,30.2,1,1,0]])
preinput
```

```python
Predict_UserInput=regressor.predict(preinput)
Predict_UserInput
```

```python
# CASE 2: 
#- Preprocess input and output from training/test data set 
#- Preprocess user input
#- Inverse preprocessed output
```

```python
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
X_train=scx.fit_transform(X_train)
X_test=scx.transform(X_test)

scy=StandardScaler()
Y_train=scy.fit_transform(Y_train)
Y_test=scy.transform(Y_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf", C=1000)
regressor=regressor.fit(X_train, Y_train)
```

```python
Y_Pred_1=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score 
r_score=r2_score(Y_test, Y_Pred_1) 
r_score
```

```python
# Preprocessing the user input before predicting the output of the user input
# Note, we are going to preprocess the output for user input and then reverse it to show it to user
preinput_1=scx.transform([[52,30.2,1,1,0]])
preinput_1
```

```python
Predict_UserInputOutput=regressor.predict(preinput_1)
Predict_UserInputOutput
```
