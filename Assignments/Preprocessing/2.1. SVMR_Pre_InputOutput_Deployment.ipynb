```python
import pickle

scx=pickle.load(open("Standardized_Input.sav", 'rb')) #Load preprocessed input
preinput=scx.transform([[52,30.2,1,1,0]]) #preprocess user input
print(preinput)

loaded_model=pickle.load(open("finalized_model_svr_1.sav", 'rb')) #load svm model
result=loaded_model.predict(preinput) #predict preprocessed user input
print(result)

#Load preprocessed output
scy=pickle.load(open("Standardized_Output.sav", 'rb')) #load preprocessed output
preoutput=scy.inverse_transform([result]) #inverse preprocessed output so the user can see original predicted value
preoutput
```
