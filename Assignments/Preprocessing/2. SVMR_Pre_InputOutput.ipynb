```python
# Importing pandas library
import pandas as pd
```

```python
# Read the dataset
dataset=pd.read_csv("insurance_pre.csv")
```

```python
# display the dataset
dataset
```

```python
# Handle categorical data
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
# Display the columns in the dataset
dataset.columns
```

```python
# Assign input values
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
# Assign output values
dependent=dataset[["charges"]]
dependent
```

```python
# Split training and test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
#Standardization
```

```python
# Applying standardization (preprocessing) for input training and test data
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
# Preprocessing input training/test data
X_train_pre=scx.fit_transform(X_train) 
X_test_pre=scx.transform(X_test) # No need to use fit_transform for preprocessing input test data since the sc variable is already holding the preprocessing logic.

scy=StandardScaler()
# Preprocessing output training/test data
Y_train_pre=scy.fit_transform(Y_train) 
Y_test_pre=scy.transform(Y_test) # No need to use fit_transform for preprocessing input test data since the sc variable is already holding the preprocessing logic.
```

```python
X_train, X_train_pre, X_test, X_test_pre
```

```python
Y_train, Y_train_pre, Y_test, Y_test_pre
```

```python
# Using SVM regression algorithm to train the Preprocessed input and output
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=1000)  
regressor.fit(X_train_pre, Y_train_pre)  # X_train & Y_train is preprocessed data 
regressor
```

```python
# Prediction based on preprocessed input test data
Y_Pred=regressor.predict(X_test_pre)
Y_Pred
```

```python
# Evaluate performance 
from sklearn.metrics import r2_score
r_score=r2_score(Y_test_pre,Y_Pred)
r_score
```

```python
# Import library to save Python objects to a file and load them 
import pickle
filename="finalized_model_svr_1.sav"
```

```python
pickle.dump(regressor, open(filename,'wb'))
```

```python
filename_scx="Standardized_Input.sav"
pickle.dump(scx, open(filename_scx,'wb'))

filename_scy="Standardized_Output.sav"
pickle.dump(scy, open(filename_scy,'wb'))
```
