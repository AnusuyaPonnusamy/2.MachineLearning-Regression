```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=50, random_state=0)
regressor.fit(X_train, Y_train)
```

```python
Y_pred=regressor.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test, Y_pred)
r_score
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_1=RandomForestRegressor(n_estimators=100, criterion='squared_error', max_depth=10, random_state=0)
regressor_1.fit(X_train, Y_train)
Y_pred_1=regressor_1.predict(X_test)
from sklearn.metrics import r2_score
r_score_1=r2_score(Y_test, Y_pred_1)
r_score_1
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_2=RandomForestRegressor(n_estimators=100, criterion='absolute_error', max_depth=10, random_state=0)
regressor_2.fit(X_train, Y_train)
Y_pred_2=regressor_2.predict(X_test)
from sklearn.metrics import r2_score
r_score_2=r2_score(Y_test, Y_pred_2)
r_score_2
```
