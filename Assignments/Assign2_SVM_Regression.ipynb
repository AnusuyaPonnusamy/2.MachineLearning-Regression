```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
dependent=dataset[["charges"]]
dependent
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="linear")  
regressor.fit(X_train, Y_train)  
```

```python
Y_Pred=regressor.predict(X_test)
```

```python
Y_Pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_Pred)
```

```python
r_score
```

```python
from sklearn.svm import SVR
regressor_rbf=SVR(kernel="rbf")  
regressor_rbf.fit(X_train, Y_train)  
```

```python
Y_Pred_rbf=regressor_rbf.predict(X_test)
Y_Pred_rbf
```

```python
from sklearn.metrics import r2_score
r_score_rbf=r2_score(Y_test,Y_Pred_rbf)
r_score_rbf
```

```python
# Penalty parameter
```

```python
from sklearn.svm import SVR
regressor_rbf=SVR(kernel="rbf", C=10)  
regressor_rbf.fit(X_train, Y_train)  
```

```python
Y_Pred_rbf=regressor_rbf.predict(X_test)
#Y_Pred_rbf
```

```python
from sklearn.metrics import r2_score
r_score_rbf=r2_score(Y_test,Y_Pred_rbf)
r_score_rbf
```

```python
from sklearn.svm import SVR
regressor_poly=SVR(kernel="poly")  
regressor_poly.fit(X_train, Y_train)  
```

```python
Y_Pred_poly=regressor_poly.predict(X_test)
Y_Pred_poly
```

```python
from sklearn.metrics import r2_score
r_score_poly=r2_score(Y_test,Y_Pred_poly)
r_score_poly
```

```python
from sklearn.svm import SVR
regressor_sig=SVR(kernel="sigmoid")  
regressor_sig.fit(X_train, Y_train)  
```

```python
Y_Pred_sig=regressor_sig.predict(X_test)
Y_Pred_sig
```

```python
from sklearn.metrics import r2_score
r_score_sig=r2_score(Y_test,Y_Pred_sig)
r_score_sig
```
